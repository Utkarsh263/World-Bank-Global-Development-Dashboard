```python
all_dfs = []

for i in range(1, 526):

    url = f"https://api.worldbank.org/v2/indicators?format=json&per_page=500&page={i}"

    response = requests.get(url)

    if response.status_code == 200:

        data = response.json()

        if len(data) < 2:
            print(f"No data at page {i}")
            continue

        indicators = data[1]

        df = pd.DataFrame([
            {
                "id": item["id"],
                "name": item["name"]
            }
            for item in indicators
        ])

        all_dfs.append(df)

        print(f"Page {i}: {len(df)} indicators collected")

    else:
        print(f"Failed to fetch page {i}, status_code={response.status_code}")
```

```python
final_df = pd.concat(all_dfs , ignore_index = True)
```

```python
indicators_groups = {

    "population": [
        "SP.POP.TOTL",        # Total Population
        "SP.DYN.LE00.IN",     # Life Expectancy at Birth
        "SP.DYN.CBRT.IN",     # Birth Rate
        "SP.DYN.CDRT.IN",     # Death Rate
        "SP.URB.TOTL.IN.ZS"   # Urban Population (%)
    ],

    "economy": [
        "NY.GDP.MKTP.CD",      # GDP (Current US$)
        "NY.GDP.PCAP.CD",      # GDP Per Capita
        "NY.GDP.MKTP.KD.ZG",   # GDP Growth (%)
        "FP.CPI.TOTL.ZG",      # Inflation (%)
        "SL.UEM.TOTL.ZS"       # Unemployment (%)
    ],

    "education": [
        "SE.ADT.LITR.ZS",      # Adult Literacy Rate
        "SE.PRM.ENRR",         # Primary School Enrollment
        "SE.SEC.ENRR",         # Secondary School Enrollment
        "SE.TER.ENRR",         # Tertiary School Enrollment
        "SE.XPD.TOTL.GD.ZS"    # Education Expenditure (% GDP)
    ],

    "health": [
        "SH.XPD.CHEX.GD.ZS",   # Health Expenditure (% GDP)
        "SH.DYN.MORT",         # Under-5 Mortality Rate
        "SH.STA.MMRT",         # Maternal Mortality Ratio
        "SH.MED.PHYS.ZS",      # Physicians per 1000 People
        "SH.IMM.MEAS"          # Measles Immunization (%)
    ],

    "environment": [
        "EN.ATM.CO2E.PC",      # CO2 Emissions per Capita
        "AG.LND.FRST.ZS",      # Forest Area (%)
        "EG.USE.PCAP.KG.OE",   # Energy Use per Capita
        "EG.ELC.ACCS.ZS",      # Access to Electricity (%)
        "ER.H2O.INTR.PC"       # Renewable Water Resources
    ],

    "technology": [
        "IT.NET.USER.ZS",      # Internet Users (%)
        "IT.CEL.SETS.P2",      # Mobile Cellular Subscriptions
        "IT.NET.BBND.P2",      # Fixed Broadband Subscriptions
        "IT.NET.SECR.P6",      # Secure Internet Servers
        "IP.JRN.ARTC.SC"       # Scientific & Technical Journal Articles
    ]

}
```

```python
import time
base_url = "https://api.worldbank.org/v2/country/all/indicator/{}?format=json&per_page=1000&page={}"
category_dataframes = {}

for category, indicators in indicators_groups.items():

    print(f"\nFetching category: {category}")

    all_dfs_for_category = []

    for indicator_code in indicators:

        print(f"Fetching indicator: {indicator_code}")

        page = 1

        while True:

            url = base_url.format(indicator_code, page)

            response = requests.get(url)

            if response.status_code != 200:
                print(f"Failed to fetch page {page} for {indicator_code}")
                break

            data = response.json()

            if len(data) < 2:
                print(f"No data found for {indicator_code}")
                break

            total_pages = data[0]["pages"]

            records = data[1]

            if not records:
                break

            df = pd.json_normalize(records)

            # Rename nested columns
            df = df.rename(columns={
                "country.id": "country_code",
                "country.value": "country_name",
                "indicator.id": "indicator_code",
                "indicator.value": "indicator_name",
                "countryiso3code": "iso3_code",
                "date": "year",
                "value": "indicator_value"
            })

            # Keep useful columns only
            useful_cols = [
                "country_code",
                "country_name",
                "iso3_code",
                "indicator_code",
                "indicator_name",
                "year",
                "indicator_value"
            ]

            available_cols = [
                col for col in useful_cols
                if col in df.columns
            ]

            df = df[available_cols]

            # Convert year to numeric
            df["year"] = pd.to_numeric(
                df["year"],
                errors="coerce"
            )

            # Keep post-2015 data
            df = df[df["year"] >= 2015]

            all_dfs_for_category.append(df)

            print(
                f"{indicator_code} | Page {page}/{total_pages} | Rows: {len(df)}"
            )

            if page >= total_pages:
                break

            page += 1

            time.sleep(0.3)

    # Combine all indicators of category

    if all_dfs_for_category:

        combined_df = pd.concat(
            all_dfs_for_category,
            ignore_index=True
        )

        category_dataframes[category] = combined_df

        print(
            f"{category} -> {len(combined_df)} total rows collected"
        )

    else:

        print(
            f"No data collected for category: {category}"
        )
```

```python
merged_dataframes = {}

for category, df in category_dataframes.items():

    temp_df = df.rename(
        columns={"iso3_code": "id"}
    )

    merged_df = pd.merge(
        temp_df,
        countries_df,
        on="id",
        how="inner"
    )

    merged_dataframes[category] = merged_df

    print(f"{category}: {merged_df.shape}")
```
